      -> Relative cell reference : when the formula changes when we copied to one cell to another cell(e.g. A1*B1 to A2*B2)
      -> Absolute cell reference : when cell doesn't change in formula when it has been copied (e.g. $A$1*$B$1  to $A$1*$B$1).
      Mixed cell reference : when we fixed either row or coloumn(e.g. $A$1*B1 to $A$1*B2)
