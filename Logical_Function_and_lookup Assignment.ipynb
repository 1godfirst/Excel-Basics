                      Logical Function And Lookup Function

1. Write a formula to check if the marks in cell B2 are greater than or equal to 40. If yes, return, "Pass", otherwise "Fail".

     -> =IF(B2>=40,"Pass","Fail")


2. company gives a bonus if sales are above 40,000 and attendance is greater than or equal to 90%. Write a formula to calculate whether the employee will get a bonus or not.

image.png

     -> =IF(AND(H10>=40000,I10>=90),"Bonus","No Bonus")

3. Write a nested IF formula:
If marks in B2 are ≥ 90 → "A", ≥ 75 → "B", ≥ 50 → "C", otherwise "Fail".

     -> =IF(B2>=90,"A",IF(B2>=75,"B",IF(B2>=50,"B","C")))
